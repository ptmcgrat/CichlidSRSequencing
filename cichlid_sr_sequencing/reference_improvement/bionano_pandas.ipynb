```python
import pandas as pd
```

```python
df = pd.read_excel('~/Desktop/bionano_summary.xlsx')
# df.columnname = [x.replace('$',' ') for x in df.columnname]
df['Avg N50 >=150kbp'] = [x.replace(' kbp', '') for x in df['Avg N50 >=150kbp']]
df['Avg N50 >=150kbp'] = df['Avg N50 >=150kbp'].astype('int')
df['Avg N50 >=20kbp'] = [x.replace(' kbp', '') for x in df['Avg N50 >=20kbp']]
df['Avg N50 >=20kbp'] = df['Avg N50 >=20kbp'].astype('int')


print(df.groupby('Tissue')['Avg N50 >=150kbp'].mean())
print('\n')
print(df.groupby('Tissue')['Avg Map Rate'].mean()*100)
print('\n')
print(df.groupby('Tissue')['Avg N50 >=20kbp'].mean())
print('\n')
print(df.groupby('Tissue')['Avg Label Density per 100kbp'].mean())


```
