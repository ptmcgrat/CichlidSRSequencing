```python
import pandas as pd, subprocess as sp
import random
```

```python
df = pd.read_excel('~/Desktop/VCF_Excel_Sheets/SampleDatabase.xlsx', sheet_name = 'vcf_samples')

# Find the counts of each unique organism
organism_counts = df['Organism'].value_counts()

# Identify organisms occurring three or more times
repeated_organisms = organism_counts[organism_counts >= 3].index.tolist()

# Create a DataFrame with a maximum of three rows for repeated organisms
max_three_df = pd.DataFrame()
for organism in df['Organism'].unique():
    organism_rows = df[df['Organism'] == organism]
    if len(organism_rows) <= 3:
        max_three_df = max_three_df.append(organism_rows)
    else:
        sampled_rows = organism_rows.sample(n=3, random_state=100)
        max_three_df = max_three_df.append(sampled_rows)

# Create a DataFrame with the remaining data (single or double occurrences)
rest_df = df[~df['Organism'].isin(repeated_organisms) | (df['Organism'].map(organism_counts) < 3)]

# Print the resulting DataFrames
print("DataFrame with a maximum of three rows for repeated organisms:")
# print(max_three_df)
max_three_df

# print("\nDataFrame with the remaining data:")
# print(rest_df)
```
