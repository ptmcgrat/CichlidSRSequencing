```python
import pandas as pd
import subprocess
```

```python
col_titles = ['Run','AvgSpotLen','Bases','BioProject','BioSample','Experiment','Instrument','Library Name','LibraryLayout','LibrarySelection','LibrarySource','Organism','Platform','SRA Study', 'FileLocations']
df = pd.DataFrame(columns=col_titles)
df
```

```python
# puts all of the sample folder names from PM24 into a list and gets rid of any standalone files and any extra whitespaces newlines
sample_files = subprocess.check_output(['rclone', 'lsf', 'ptm_dropbox:/CoS/BioSci/BioSci-McGrath/Apps/CichlidSequencingData/Reads/PM24', '--dirs-only'], encoding = 'utf-8').strip().split('\n')
# sample_files
```

```python
# Use this to verify that the number of dirs found is how many there should be!!
len(sample_files)
```

```python
run = []
concise_samples = []
for sample in sample_files:
    concise_samples.append(sample.split('.')[0][:-3])
len(run)
df['Run'] = concise_samples
```

```python
# defining lots of the same vvariables here.
AvgSpotLen = '300'
Bases= '0'
BioProject = 'bionano_paper'
Experiment = 'bionano_paper'
Instrument = 'Illumina NovaSeq 6000'
LibraryLayout = 'PAIRED'
LibrarySelection = '-'
LibrarySource = 'GENOMIC'
Platform = 'ILLUMINA' 
SRA_Study = 'SexDetermination'
```

```python
df['AvgSpotLen'] = [AvgSpotLen]*len(df)
df['Bases'] = [Bases]*len(df)
df['BioProject'] = [BioProject]*len(df)
df['Experiment'] = [Experiment]*len(df)
df['Instrument'] = [Instrument]*len(df)
df['LibraryLayout'] = [LibraryLayout]*len(df)
df['LibrarySelection'] = [LibrarySelection]*len(df)
df['LibrarySource'] = [LibrarySource]*len(df)
df['Platform'] = [Platform]*len(df)
df['SRA Study'] = [SRA_Study]*len(df)
```

```python
# to remove 'lab_' prefixes and '_L001' and '_L002' suffixes
df['BioSample'] = concise_samples # from above
df['BioSample'] = df.BioSample.replace('lab_', '', regex=True)
df['BioSample'] = df.BioSample.replace('_L001', '', regex=True)
df['BioSample'] = df.BioSample.replace('_L002', '', regex=True)
```

```python
# add the 'Library Name' Column

Library_Name = df['BioSample'] + '_library'
df['Library Name'] = Library_Name
df
```

```python
# Add the FileLocations
# also removes any non fastq files from the paths and prints length to confirm the number of paths you have
paths = subprocess.check_output(['rclone', 'lsf', 'ptm_dropbox:/CoS/BioSci/BioSci-McGrath/Apps/CichlidSequencingData/Reads/PM24', '--format', 'p', '--files-only', '-R'], encoding = 'utf-8').strip().split('\n')
paths = [x for x in paths if x.endswith('.fastq.gz')]
print('Number of FASTQ files is: ' + str(len(paths)))
```

```python
paths
concat_paths = []
for path1,path2 in zip(*[iter(paths)]*2):
    sample_paths = []
    sample_paths.append('PM24/'+ path1)
    sample_paths.append('PM24/' + path2)
    joined_paths = ',,'.join(sample_paths)
    concat_paths.append(joined_paths)
concat_paths[:3]
```

```python
# Do the complicated, yet not so complicated filepath mapping magic
# Be sure to check each sample to make sure that this is correct before running the pipelines!!!!!!!!
for sample in concise_samples:
    for path in concat_paths:
        if sample in path:
            df.loc[df.Run == sample, 'FileLocations'] = path
```
